### fréquences de caractères
Le code suivant permet de visualiser la fréquence des caractères présents dans le texte.

```python
import pandas as pd
import matplotlib.pyplot as plt

NATURAL_TEXT_PATH = "./assets/code3.css"
f= open(NATURAL_TEXT_PATH, "r")

textString = f.read()

data = {}

for char in textString :
    if(data.get(char)) :
        data[char] += 1
    else : 
        data.setdefault(char, 1)

print(data)

plt.bar(data.keys(), data.values())

plt.title("frequences des characteres")

plt.show()
```

### fréquences de groupes de characteres (taille > 2 chars)
Le code suivant permet de visualiser la fréquence des groupes de characteres présents dans le texte. (avec max length = 20)

```python
NATURAL_TEXT_PATH = "./assets/code3.css"
f= open(NATURAL_TEXT_PATH, "r")

textString = f.read()

charDict = {}

sampleSize = 2

charDictFiltered = {}

while sampleSize < 20 :
    i = 0
    while i + sampleSize <= len(textString)  :
        sample = textString[i:i+sampleSize]

        if (not sample in charDict) : 
            charDict.setdefault(sample, (sampleSize, 1))
        else : 
    
            charDict[sample] = (sampleSize, charDict[sample][1] + 1)
        
        i += 1
    
    sampleSize += 1

    

    for key in charDict :
        if charDict[key][1] >= 2 :
            charDictFiltered.setdefault(key, charDict[key])

sizeCharDict = {}

for key in charDictFiltered :

   

    if (not charDictFiltered[key][0] in sizeCharDict) : 
        sizeCharDict.setdefault(charDictFiltered[key][0], charDictFiltered[key][1])
    else : 
        sizeCharDict[charDictFiltered[key][0]] += charDictFiltered[key][1]


print(sizeCharDict)

plt.bar(sizeCharDict.keys(), sizeCharDict.values())
plt.title("frequences des groupes de characteres")
plt.xticks(range(0, 21))
plt.show()
```
